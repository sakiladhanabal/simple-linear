```python
#pandas is a library and it contains the function related to the database
import pandas as pd
```

```python
#dataset is called read_csv (salary data csv ).pd database
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset# type dataset and get the information of data set(model creation first step datacollection)
```

```python
independent=dataset[["YearsExperience"]]#split the input data from dataset this problem dataset called as years of expreience
independent
```

```python
#this is splited output dataset(salary)
dependent=dataset[["Salary"]]
dependent
```

```python
#from sklearn library modelselection function is called and import splited train and test dataset assign
#variable for splited train and test datas
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30,random_state=0)
```

```python
#from sklearn library linearmodel function is called and import linear regression
#assign variable name regressor to linearregression and use fitfunction in x,y trainset
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#weight function is checked by coef_ variable
weight=regressor.coef_
weight
```

```python
#bais function is checked by intercept_ variable
bais=regressor.intercept_
bais
```

```python
#create y_pred variable cross checked the prediction of x_test evalution metrics
#include train set  value in test set value
y_pred=regressor.predict(x_test)
```

```python
#create r2_score named variable pass actual output and predicted output to r2_score variable find evalution metrics
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#find r_score if it is nearly 1 it is good model
r_score
```

```python
#use pickle for save file
import pickle 
filename="finalized_model_linear.sav"
```

```python
#'wb ' open the file and regressor model dump in pickle
pickle.dump(regressor,open(filename,'wb'))
```

```python
#after save file load the model and predict given the model find the result
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
```
